```python
from sklearn.datasets import load_iris
import pandas as pd

# โหลดชุดข้อมูล Iris
iris = load_iris()

# แปลงเป็น DataFrame ของ pandas
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target  # เพิ่มคอลัมน์ target (species)
```

```python
# แสดงข้อมูลบางส่วนแรก
print(iris_df.head())

# สถิติสรุป
print(iris_df.describe())

# ตรวจสอบค่าที่หายไป
print(iris_df.isnull().sum())

# นับจำนวน species ที่ไม่ซ้ำกัน
print(iris_df['species'].value_counts())
```

```python
iris_df.to_csv('iris_dataset.csv', index=False)
```
